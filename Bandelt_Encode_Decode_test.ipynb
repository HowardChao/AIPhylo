```python
import os
import sys
```

```python
import Bandelt_Encode_Decode.Bandelt_Encode_func as BN_Encode_func
import Bandelt_Encode_Decode.Bandelt_Decode_func as BN_Decode_func
import Bandelt_Encode_Decode.Bandelt_Node as BN
```

# Get mapping_dic_dic first

```python
folder_list = ['Bergsten_2013', 'Broughton_2013']
# folder_list = ['Bergsten_2013', 'Cannon_2016_dna', 'Leache_2015', 'Prebus_2017', 'Smith_2014', 'Broughton_2013', 'Crawford_2012', 'McCormack_2013', 'Ran_2018_dna', 'Wu_2018_dna', 'Brown_2012', 'Faircloth_2013', 'Meiklejohn_2016', 'Richart_2015']
```

```python
training_dataset_prefix = "training_small"
# training_dataset_prefix = "training_big"
```

```python
dic={'A':'1', 'T':'2', 'C':'3', 'G':'4', '-':'0', '?':'0'}
```

```python
mapping_dic_dic = {}
mapping_dic_dic_decode = {}
```

```python
# data preprocessing
for folder in folder_list:
    file_path = os.path.join(training_dataset_prefix, folder)
    files_list = os.listdir(file_path)
    for file in files_list:
        if file.find('.nex.treefile.dist') > 0:
            file_base_name = file[0:file.find('.nex.treefile.dist')]
            #######################
            # Input Processing !! #
            #######################
            print(os.path.join(file_path, file_base_name+'.nex'))
            seq_data_raw = open(os.path.join(file_path, file_base_name+'.nex'))
            seq_data = seq_data_raw.readlines()[6:]
            species_mapping_dic = {}
            species_mapping_dic_decode = {}
            for idx, line in enumerate(seq_data):
                curr_line = line.split()
                if len(curr_line) < 2:
                    break
#                 print(curr_line[0])
                if int(idx) == 0:
                    species_mapping_dic[curr_line[0]] = sys.maxsize
                    species_mapping_dic_decode[sys.maxsize] = curr_line[0]
                else: 
                    species_mapping_dic[curr_line[0]] = int(idx-1)
                    species_mapping_dic_decode[int(idx-1)] = curr_line[0]
            mapping_dic_dic[os.path.join(file_path, file_base_name+'.nex')] = species_mapping_dic
            mapping_dic_dic_decode[os.path.join(file_path, file_base_name+'.nex')] = species_mapping_dic_decode
```

# Recursively check every files

```python
for folder in folder_list:
    file_path = os.path.join(training_dataset_prefix, folder)
    files_list = os.listdir(file_path)
    for file in files_list:
        if file.find('.nex.treefile.dist') > 0:
            file_base_name = file[0:file.find('.nex.treefile.dist')]
            BN_encode_tree, BN_encode_list = BN_Encode_func.Bandelt_Encode(file_path, file_base_name, mapping_dic_dic)
            BN_decode_tree, BN_decode_newick = BN_Decode_func.Bandelt_Decode(BN_encode_list, file_path, file_base_name, mapping_dic_dic_decode)
            print('File: ', os.path.join(file_path, file_base_name+'.nex'))
            print("************ Encode Tree!! ************")
            BN_encode_tree.print_subtree()
            print('>>>> BN_encode_list: ', BN_encode_list)
            print("************ Decode Tree!! ************")
            BN_decode_tree.print_subtree()
            print('>>>> BN_decode_newick: ', BN_decode_newick)
            print('Encode and decode tree compare: ', BN_decode_tree.compare_subtree(BN_encode_tree))
            print('=======================================')
            print('\n\n')
```
