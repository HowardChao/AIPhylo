```python
training_dataset_prefix = "training_big"
```

```python
BRANCH_NUM = 100
SEQUENCE_LEN = 1000
```

```python
mapping_dic_dic = {}
mapping_dic_dic_decode = {}
```

```python
X_Input_Alignment_List = []
Y_Output_BN_List = []
```

```python
# data preprocessing
for folder in folder_list:
    file_path = os.path.join(training_dataset_prefix, folder)
    files_list = os.listdir(file_path)
    for file in files_list:
        if file.find('.nex.treefile.dist') > 0:
            file_base_name = file[0:file.find('.nex.treefile.dist')]
            print(file_base_name)            

            #######################
            # Input Processing !! #
            #######################
            print(os.path.join(file_path, file_base_name+'.nex'))
            alignment_data_raw = open(os.path.join(file_path, file_base_name+'.nex'))            
            alignment_data = alignment_data_raw.readlines()[6:]
            
            ## This two dictionary are for output data encoding/decoding
            species_mapping_dic = {}
            species_mapping_dic_decode = {}
            
            ## This is one X_Input_Alignment which will be replaced in each iteration
            X_Input_Alignment = np.zeros((BRANCH_NUM,SEQUENCE_LEN,5))
            for idx, line in enumerate(alignment_data):
                curr_line = line.split()
                if len(curr_line) < 2:
                    break
                    
                ## Output data encoding / decoding dictionary
                if int(idx) == 0:
                    species_mapping_dic[curr_line[0]] = sys.maxsize
                    species_mapping_dic_decode[sys.maxsize] = curr_line[0]
                else: 
                    species_mapping_dic[curr_line[0]] = int(idx-1)
                    species_mapping_dic_decode[int(idx-1)] = curr_line[0]
                
                ## Input alignment preprocessing
                alignment_data_str = curr_line[1]
                for idx_inner, each_string in enumerate(alignment_data_str):
                    if each_string == 'A':
                        X_Input_Alignment[idx, idx_inner] = np.array([1, 0, 0, 0, 0])
                    elif each_string == 'T':
                        X_Input_Alignment[idx, idx_inner] = np.array([0, 1, 0, 0, 0])
                    elif each_string == 'C':
                        X_Input_Alignment[idx, idx_inner] = np.array([0, 0, 1, 0, 0])
                    elif each_string == 'G':
                        X_Input_Alignment[idx, idx_inner] = np.array([0, 0, 0, 1, 0])
                    else:
                        X_Input_Alignment[idx, idx_inner] = np.array([0, 0, 0, 0, 1])
                
            ## Output data encoding / decoding dictionary  
            mapping_dic_dic[os.path.join(file_path, file_base_name+'.nex')] = species_mapping_dic
            mapping_dic_dic_decode[os.path.join(file_path, file_base_name+'.nex')] = species_mapping_dic_decode
            
            ## Input alignment preprocessing
            X_Input_Alignment_List.append(X_Input_Alignment)
#             print('X_Input_Alignment.shape: ', X_Input_Alignment.shape)

            ########################
            # Output Processing !! #
            ########################
            BN_encode_tree, BN_encode_list = BN_Encode_func.Bandelt_Encode(file_path, file_base_name, mapping_dic_dic)
            BN_decode_tree, BN_decode_newick = BN_Decode_func.Bandelt_Decode(BN_encode_list, file_path, file_base_name, mapping_dic_dic_decode)
            newick_decode_file = open(os.path.join('Data_Preprocessing/Bandelt_Input_output_preprocessing/output/28630_Dataset', folder, file_base_name+'_encode.BN'), "w")
            n = newick_decode_file.write(BN_decode_newick)
            newick_decode_file.close()    
            
```
