```python
# folder_list = ['Bergsten_2013', 
#                'Broughton_2013',
#                'Brown_2012', 
#                'Cannon_2016_dna',
#                'Crawford_2012', 
#                'Faircloth_2013', 
#                'Leache_2015',
#                'McCormack_2013', 
#                'Meiklejohn_2016', 
#                'Prebus_2017', 
#                'Ran_2018_dna', 
#                'Richart_2015', 
#                'Smith_2014',
#                'Wu_2018_dna']

folder_list = ['Bergsten_2013', 'Broughton_2013', 'Brown_2012', 'Cannon_2016_dna']
```

```python
dataset_prefix = "training_big"
```

```python
BRANCH_NUM = 100
SEQUENCE_LEN = 1000
```

```python
with open('./ALL_DATA/mapping_dic_dic.npy', 'rb') as input:
    mapping_dic_dic = pickle.load(input)
    
with open('./ALL_DATA/mapping_dic_dic_decode.npy', 'rb') as input:
    mapping_dic_dic_decode = pickle.load(input)
```

```python
X_Input_Alignment_List = []
Y_Output_BN_List = []
```

```python
FILE_NUM = 0
```

```python
# data preprocessing
for folder in folder_list:
    file_path = os.path.join(dataset_prefix, folder)
    files_list = os.listdir(file_path)
    for file in files_list:
        if file.find('.nex.treefile.dist') > 0:
            FILE_NUM = FILE_NUM + 1
            
            file_base_name = file[0:file.find('.nex.treefile.dist')]
            print(file_base_name)            

            #######################
            # Input Processing !! #
            #######################
            print(os.path.join(file_path, file_base_name+'.nex'))
            alignment_data_raw = open(os.path.join(file_path, file_base_name+'.nex'))            
            alignment_data = alignment_data_raw.readlines()[6:]
            
            ## This is one X_Input_Alignment which will be replaced in each iteration
            X_Input_Alignment = np.zeros((BRANCH_NUM,SEQUENCE_LEN,5))
            for idx, line in enumerate(alignment_data):
                curr_line = line.split()
                if len(curr_line) < 2:
                    break
                
                ## Input alignment preprocessing
                alignment_data_str = curr_line[1]
                for idx_inner, each_string in enumerate(alignment_data_str):
                    if idx_inner >= 1000:
                        continue
                    if each_string == 'A':
                        X_Input_Alignment[idx, idx_inner] = np.array([1, 0, 0, 0, 0])
                    elif each_string == 'T':
                        X_Input_Alignment[idx, idx_inner] = np.array([0, 1, 0, 0, 0])
                    elif each_string == 'C':
                        X_Input_Alignment[idx, idx_inner] = np.array([0, 0, 1, 0, 0])
                    elif each_string == 'G':
                        X_Input_Alignment[idx, idx_inner] = np.array([0, 0, 0, 1, 0])
                    else:
                        X_Input_Alignment[idx, idx_inner] = np.array([0, 0, 0, 0, 1])
            
            ## Input alignment preprocessing
            X_Input_Alignment_List.append(X_Input_Alignment)

            ########################
            # Output Processing !! #
            ########################
            BN_encode_list = BN_Encode_func.Bandelt_Encode(file_path, file_base_name, mapping_dic_dic)
            with open(os.path.join('ALL_DATA/Bandelt_Encoding_Decoding_Raw_File/Encoding/28630_Dataset', folder, file_base_name+'_bn_encode.pkl'), 'rb') as input:
                BN_encode_list = pickle.load(input)
            ## Pop out the first element because it must be 0!!
            BN_encode_list.pop(0)
            ## Add zero padding so that all output length are same!!
            BN_encode_list.extend([0] * (BRANCH_NUM - len(BN_encode_list)))
            Y_Output_BN_List.append(BN_encode_list)
```

```python
X_Input_Alignment_Data = np.stack(X_Input_Alignment_List,axis=0)
Y_Output_BN_List_Data = np.stack(Y_Output_BN_List,axis=0)

print('X_Input_Alignment_Data.shape: ', X_Input_Alignment_Data.shape)
print('Y_Output_BN_List_Data.shape: ', Y_Output_BN_List_Data.shape)
```

```python
np.save('./ALL_DATA/499_dataset_Input_Output/X_Input_Alignment_Data.npy', X_Input_Alignment_Data)
np.save('./ALL_DATA/499_dataset_Input_Output/Y_Output_BN_List_Data.npy', Y_Output_BN_List_Data)
```
